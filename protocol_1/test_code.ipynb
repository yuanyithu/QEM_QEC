```python
L_list = np.linspace(1,1001,10,dtype=int)
p = 0.001

def sample(p,L):
    pL= (1-(1-4*p)**L)/4
    return ((1+2*pL)/(1-4*pL))**2

from matplotlib import pyplot as plt
import numpy as np
result = []
for i in range(len(L_list)):
    result.append(np.log10(sample(p,L_list[i])))

plt.plot(L_list,result)
plt.show()
```

$N(\rho) = (1-3p)I\rho I + p (X\rho X+Y \rho Y+Z\rho Z)$

$N(X) = (1-4p)X$

$N^{-1}(\rho) = \frac{1}{1-4p}\rho-$

$p_L = \frac{1-(1-4p)^L}{4}$

$sample = \frac{1+2p_L}{1-4p_L}$
